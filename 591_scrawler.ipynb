```python
import requests
from bs4 import BeautifulSoup
rs  = requests.session()
cookies = {'urlJumpIp': '1'}
res = rs.get('https://rent.591.com.tw/?kind=0&region=1&shType=clinch',cookies=cookies)
soup = BeautifulSoup(res.text,'lxml')
token = soup.select_one('meta[name="csrf-token"]')['content']
```

```python
headers={
    'accept-encoding':"utf-8",
    'X-CSRF-TOKEN':token,
    'X-Requested-With':'XMLHttpRequest'
}

apiurl='https://rent.591.com.tw/home/search/rsList?is_new_list=1&type=1&kind=&searchtype=1&region=1&shType=clinch'

res=rs.get(apiurl,headers=headers,cookies = {'urlJumpIp': '1'})
data = res.json()
```
